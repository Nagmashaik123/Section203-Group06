```python
# We have fetched 10000 records per one user API request. 
# Called 
# Concatinating them to one single file as #CovidVaccine.csv

import os
import glob
import pandas as pd
os.chdir("/content")

extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

#combine all files in the list
dataFrame = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
dataFrame.to_csv( "CovidVaccineWorldWide.csv", index=False, encoding='utf-8-sig')
dataFrame
```
